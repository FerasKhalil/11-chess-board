```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class ChessBoard():
    def __init__(self):
        self.grid = np.zeros((8,8,3))
        self.red_position=None
        self.blue_position=None

        white = (1,1,1)
        for i in range (8):
            for j in range(8):
                if (i%2==0 and j%2==0) or (i%2!=0 and j%2!=0):
                    self.grid[i,j]=white

    def add_red(self,row,col):
        red = (1,.2,0)
        self.grid[row,col]=red
        self.red_position=(row,col)
    
    def add_blue(self,row,col):
        blue = (0,1,1)
        self.grid[row,col]=blue
        self.blue_position=(row,col)
    
    def render(self):
        plt.imshow(self.grid)

    def is_under_attack(self):
        red=self.red_position
        blue=self.blue_position
        if red[0]==blue[0] or red[1]==blue[1] or (red[0]-blue[0]==red[1]-blue[1]):
            return True
        else:
            return False
```

```python
chess=ChessBoard()
chess.render(4,5,1,5)
print(chess.is_under_attack())
```

```python
def test_same_raw():
    chess=ChessBoard()
    chess.render(4,1,4,3)
    assert chess.is_under_attack()==True
    print('under attack')
    
test_same_raw()
```

```python

def test_same_column_under_attack():
    chess=ChessBoard()
    chess.render(0,1,6,1)
    assert chess.is_under_attack()==True
    print('under attack')
    
test_same_column_under_attack()
```

```python

def test_same_diagonal_under_attack():
    chess=ChessBoard()
    chess.render(2,1,4,3)
    assert chess.is_under_attack()==True
    print('under attack')
    
test_same_diagonal_under_attack()
```
